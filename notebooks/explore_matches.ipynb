```python
import json
import pandas as pd
```

```python
with open('../data/raw/matches_20260512_085115.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
partidos = list()

for match in data:
    partidos.append({
        "match_id": match['id'],
        "country": match['area']['name'],
        "competition": match['competition']['name'],
        "utc_date": match['utcDate'],
        "status": match['status'],
        "season": f"{match['season']['startDate'][:4]} - {match['season']['endDate'][:4]}",
        "matchday": match['matchday'],
        "home_team": match['homeTeam']['name'],
        "away_team": match['awayTeam']['name'],
        "home_goals": match['score']['fullTime']['home'],
        "away_goals": match['score']['fullTime']['away'],
        "winner": match['score']['winner'],
        "referee": match['referees'][0]['name'] if match['referees'] else "Desconocido"
    })
```

```python
df = pd.DataFrame(partidos)

# Partidos finalizados
df_new = df.loc[df["status"] == 'FINISHED']
df_new["home_goals"] = df_new["home_goals"].astype(int)
df_new["away_goals"] = df_new["away_goals"].astype(int)
df_new["utc_date"] = pd.to_datetime(df_new["utc_date"])
```
